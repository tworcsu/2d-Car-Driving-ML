## Introduction

The goal of the project is to teach a car to drive in a two dimensional setting across a hilly train.  Similarly to shown bellow from a online program [boxcar2d](http://boxcar2d.com/).  This project will be similar in the goal of making the car travel as far as possible on the terrain before becoming stuck.  Although it is different because instead of trying to make new cars to travel accross the terrain, the program will change the way the car drives across the terrain to try to go as far as possible.  The only control the car has is the throttle.  If the car goes to quick up a hill this will cause the car to flip over which results in an immediate failure.  If it goes to slowly it makes a bad time or may even fail to go up a hill.

![](https://i.ytimg.com/vi/lO6OWqOQmM4/hqdefault.jpg)

This makes the project suitable for using reinforcement learning because there is a clear way to measure success.  *how far has the car gone and the time it made*

Likely the most difficult part of this project will be to synthesize the terrain and be able to read the terrain into a training input that is useful.  For example, the height of the terrain is not so important at any point.  What is more imporant is how high the terrain at some point relative to the neighbors around it.

There are a few ways of tacking this issue.  One of which is to interpret the terrain into WaveLets or using a Fourier Transform.  Another way to gather meaningful input from the raw terrain data is to have several height inputs for the nearest points in the terrain directly in front of the car that are relative to the car's current height.

For comparison, we will use two dumb algorithms that do not use machine learning in any way.  One will just go at a constant speed.  It can be considered to be the control.  The other will simply look at the terrain ahead.  If the terrain is sloping upward is will push forward if not, it will just coast unless it is not moving already.

## Methods

For this project, we will be using the reinforcement learning algorthms discussed in class.  The data the car will learn with will be generated by combining several sin waves on top of each other along with randomness.  The sin waves will become more irregular as the terrain becomes longer.  

## Possible Results

The dumb algorthms should fail very quickly.  When the inputs are processed through wavlets, we expect the car will travel further and faster.  This is because the Wavelets might be able to better help the reinforcement learning algorithm understand its terrain.
